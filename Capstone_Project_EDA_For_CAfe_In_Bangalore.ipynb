```python
#importing important libraries rquired for us 
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the data set
df = pd.read_csv("/content/drive/MyDrive/Data Science Files/dineout1 (6).csv")
df.head()
```

```python
#deleting 1st 3 columns we dont want
df = df.drop(df.columns[[0, 1, 2, 13 ,14 ,16 ,17]], axis=1)  # df.columns is zero-based pd.Index
```

```python
#rename the column
df.rename(columns = {'reat_name':'resto_name'}, inplace = True)
```

```python
#displaying dataset after deleting unwanted data
df.head()
```

```python
#checking the shape of our data set
df.shape
```

```python
#taking info of our dataset
df.info()
```

```python
#checking the null values in our dataset
df.isnull().sum()
```

```python
#checking unique values in our dataset
df.nunique()
```

```python
unique_df = pd.DataFrame()
unique_df['Features'] = df.columns
unique=[]
for i in df.columns:
    unique.append(df[i].nunique())
unique_df['Uniques'] = unique

f, ax = plt.subplots(1,1, figsize=(15,7))

splot = sns.barplot(x=unique_df['Features'], y=unique_df['Uniques'], alpha=1)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center',
                   va = 'center', xytext = (0, 9), textcoords = 'offset points')
plt.title('Bar plot for number of unique values in each column',weight='bold', size=15)
plt.ylabel('#Unique values', size=12, weight='bold')
plt.xlabel('Features', size=12, weight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
#for the count of rest. area wise we groupby addres and count of rest. reat_name
df1 = df.groupby(['address'])['resto_name'].aggregate('count').reset_index().sort_values('resto_name', ascending=False)
df1.rename(columns = {'resto_name':'Number_of_resto'}, inplace = True)
```

```python
#info of df1
df1.info()
```

```python
#displaying the values in df1
display(df1)
```

```python
#convert the avg_cost in numeric form 
df3['avg_cost'] = df3['avg_cost'].str.rstrip('for two people')
df3['avg_cost'] = df3['avg_cost'].str.replace(',', '')

#rename the columns name avg_cost to avg_cost_for_two
df3.rename(columns = {'avg_cost':'avg_cost_for_two'}, inplace = True)
```

```python
# convert "avg_cost_for_two" from String to int
df3 = df3.astype({'avg_cost_for_two':'int64'})

#other ways to convert

#df3['avg_cost_for_two'].astype(object).astype(int)

#pd.to_numeric(df3['avg_cost_for_two'])
```

```python
#for count of area wise avg cost in bangalore.
df6 = df3.groupby(['address'])['avg_cost_for_two',].aggregate('median').reset_index().sort_values('avg_cost_for_two', ascending=False)
```

```python
df6.head()
```

```python
#ploting graph bet count vs type
fig , ax = plt.subplots(figsize = (20,8))
splot= sns.barplot(x = 'address', y = 'avg_cost_for_two' , data = df6.head(10))

# now simply assign the bar values to
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.1f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center',
                   va = 'center', xytext = (0, 9), textcoords = 'offset points')
plt.xlabel('Address',size = 20,weight='bold')
plt.ylabel('Avg Cost For Two',size = 20 , weight = 'bold')
plt.xticks(rotation = 25)

```
